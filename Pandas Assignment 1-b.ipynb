#### #2. Read in the NBA players `csv` into a variable called `nba_df`.

This is a data set of NBA players from 2015. The filename is `NBA_players_2015.csv`.

```python
df = pd.read_csv('NBA_players_2015.csv', header=None)
nba_df = df[0].str.split(',',expand= True)
nba_df.columns = nba_df.loc[0]
nba_df = nba_df[1:]
```

#### #6. Sort the data set in on the `players` column in alphabetical order.

```python
nba_df.sort_values(by=('player'), inplace=True)
nba_df.head()
```

#### #7. Filter the data set. Create three sub DataFrames from the `position` column for `G`, `F`, and `C`.

```python
g_df = nba_df[nba_df['pos'] == 'G']
f_df = nba_df[nba_df['pos'] == 'F']
c_df = nba_df[nba_df['pos'] == 'C']

print('Guards:')
print(g_df.head())
print('\nForwards:')
print(f_df.head())
print('\nCenters:')
print(c_df.head())
```

#### #8.  Run `describe()` on these new DataFrames. Compare the mean field goals (the `fg` column) between positions.

```python
g_df.describe()
g_fg_mean = g_df['fg'].astype(float).mean()
g_fg_mean
```

```python
f_df.describe()
f_fg_mean = f_df['fg'].astype(float).mean()
f_fg_mean
```

```python
c_df.describe()
c_fg_mean = c_df['fg'].astype(float).mean()
c_fg_mean
```
